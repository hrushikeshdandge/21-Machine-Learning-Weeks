```
import keras

# x_train and y_train are Numpy arrays --just like in the Scikit-Learn API.
from keras.datasets import mnist
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split

df=load_digits()
x_train,x_test,y_train,y_test=train_test_split(df.data,df.target)

#(x_train, y_train), (x_test, y_test) = mnist.load_data()

from keras.utils import to_categorical

y_train=to_categorical(y_train, num_classes=None)
y_test=to_categorical(y_test, num_classes=None)


# x_train=x_train.reshape(x_train.shape[0],x_train.shape[1]*x_train.shape[2])
# x_test=x_test.reshape(x_test.shape[0],x_test.shape[1]*x_test.shape[2])


#building model
from keras.models import Sequential
model = Sequential()
from keras.layers import Dense

n_features=x_train.shape[1]
n_labels=y_train.shape[1]

print(x_train.shape)
print(y_train.shape)

model.add(Dense(units=64, activation='relu', input_dim=n_features))
model.add(Dense(units=n_labels, activation='softmax'))


model.compile(loss='categorical_crossentropy',
              optimizer='sgd',
              metrics=['accuracy'])

model.fit(x_train, y_train, epochs=30, batch_size=20,shuffle=True)

# from keras.utils import multi_gpu_model
# multi_gpu_model(model, gpus=1)
# parallel_model = multi_gpu_model(model, gpus=8)
# parallel_model.compile(loss='categorical_crossentropy',
#                        optimizer='adam')

#parallel_model.fit(x, y, epochs=20, batch_size=256)

# Save model via the template model (which shares the same weights):
model.save('my_model.h5')

loss_and_metrics = model.evaluate(x_test, y_test, batch_size=32)
print(loss_and_metrics)

print("done")

```
